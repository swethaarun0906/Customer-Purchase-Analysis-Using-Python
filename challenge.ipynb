```python
class InvalidAmountError(Exception):
    try:
        amount = float(input("Enter transaction amount: "))
        if amount <= 0:
            raise InvalidAmountError()
            print("Transaction amount not valid.")
        else:
            print("Transaction amount valid.")
    except InvalidAmountError as e:
        print("Error: Invalid Transaction amount.Enter a positive Value")
    except ValueError:
        print("Error: Please enter a valid number.")
```

```python
def car_features(**args):
    print("Car Features:")
    for key, value in args.items():
        print(f"{key}: {value}")
count = int(input("How many features? "))
features = {}
for i in range(count):
    name = input("Enter feature name: ")
    value = input("Enter feature value: ")
    features[name] = value


car_features(**features)
```
