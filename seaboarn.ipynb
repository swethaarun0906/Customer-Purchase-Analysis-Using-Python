```python
import pandas as pd
data={"Name":["swe","krish","pranav","sneha","mani"],"Maths":[80,60,50,40,55],"science":[70,90,80,75,60],"Gender":["female","male","male","female","female"]}
print(data)
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
sns.barplot(x="Name",y="Maths", data=df, palette="Greens")
plt.title("SEABORN")
plt.xticks(rotation=50)
```

```python
sns.lineplot(x="Name",y="science",data=df, marker="s", color="red")
plt.title("LINEPLOT")
```

```python
sns.boxplot(x="Gender",y="Maths", data=df)
plt.title("BOXPLOT")
plt.xticks(rotation=50)
```

```python
sns.violinplot(x="Gender",y="Maths", data=df)
plt.title("VIOLPLOT")
plt.xticks(rotation=50)
```

```python
sns.scatterplot(x="Name",y="Maths", hue="Gender",data=df)
```

```python
sns.heatmap(df[["Maths","science"]].corr(),annot=True,cmap='pink')
```
