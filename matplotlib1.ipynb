```python
import pandas as pd
data={"Name":["swe","krish","pranav","sneha","mani"],"Maths":[80,60,50,40,55],"science":[70,90,80,75,60],"Gender":["female","male","male","female","female"]}
print(data)
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(df["Name"],df["Maths"], color="purple")
plt.title("class")
plt.xticks(rotation=45)
plt.xlabel("Name")
plt.ylabel("Maths")
plt.show()
```

```python
plt.plot(df["Name"],df["science"],color="red",linestyle="dotted",linewidth=5,marker="^",markersize="15")
plt.xticks(rotation=45)
plt.grid(True)
plt.ylabel("science")
plt.xlabel("Name")
plt.title("class")
```

```python
plt.hist(df["Maths"], bins=5,edgecolor="black")
```

```python
plt.scatter(df["science"],df["Maths"], color="orange", edgecolor="black")
plt.title("Subject")

plt.xlabel("science")
plt.ylabel("Maths")
plt.show()
```
